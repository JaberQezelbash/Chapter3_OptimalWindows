```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

# Original data
x_values = [600, 720, 960, 1200, 1450]

# Metrics data
metrics = {
    'Accuracy': [79, 80, 83, 90, 99],
    'Precision': [42, 50, 63, 77, 99],
    'Recall': [56, 59, 65, 75, 99],
    'Specificity': [73, 76, 81, 89, 99]
}

# Function to smooth the data using spline interpolation
def smooth_curve(x, y, num_points=300):
    # Convert lists to numpy arrays for processing
    x = np.array(x)
    y = np.array(y)
    
    # Generate new x values for smoothness
    x_new = np.linspace(x.min(), x.max(), num_points)
    
    # Create a spline of degree 3 (cubic)
    spline = make_interp_spline(x, y, k=3)
    y_new = spline(x_new)
    
    return x_new, y_new

# Create a new figure with a specified smaller size
plt.figure(figsize=(7.2, 5.5))  # Changed from (12, 8) to (8, 6)

# Define a color palette for distinct curves
colors = ['blue', 'green', 'red', 'purple']
markers = ['o', 's', '^', 'D']  # Different markers for each curve

# Plot each metric
for idx, (metric, y_values) in enumerate(metrics.items()):
    # Smooth the curve
    x_smooth, y_smooth = smooth_curve(x_values, y_values)
    
    # Plot the smoothed curve
    plt.plot(
        x_smooth, y_smooth,
        label=metric,
        color=colors[idx],
        linewidth=2,
        linestyle='-',
        marker=None  # No markers on the smoothed line
    )
    
    # Plot the original data points without labels to exclude them from the legend
    plt.scatter(
        x_values, y_values,
        color=colors[idx],
        marker=markers[idx],
        edgecolor='k',
        s=60,  # Size of the markers
        label=None  # Remove label to exclude from legend
    )

# Add title and labels (Uncomment and customize the title if needed)
# plt.title('Performance Metrics vs. Window Size', fontsize=12, fontname='Arial', fontweight='bold')

plt.xlabel('Windows Size', fontsize=12, fontname='Arial')
plt.ylabel('Performance Metrics (%)', fontsize=12, fontname='Arial')

# Add grid for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

# Customize ticks to show only the specified x_values
plt.xticks(x_values, fontsize=12, fontname='Arial')  # Set x-ticks to specific values
plt.yticks(fontsize=12, fontname='Arial')

# Add legend with Arial font and size 12
plt.legend(fontsize=12, loc='lower right', prop={'family': 'Arial', 'size': 12})

# Tight layout for better spacing
plt.tight_layout()

# Save the figure as a PDF
plt.savefig('performance_metrics_plot.pdf', format='pdf')

# Save the figure as an SVG
plt.savefig('performance_metrics_plot.svg', format='svg')

# Optional: Display the plot
plt.show()
```
